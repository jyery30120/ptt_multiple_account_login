```python
import telnetlib
import sys
import time


host = 'ptt.cc'
users = ['username_1','username_2','username_3'] ## change here , can be only one or add more than 1
passwords = ['userpassword_1','userpassword_2','userpassword_3'] ## change here

## users = ['username_1'] ## change here , can be only one or add more than 1
## passwords = ['1234'] ## change here


accounts=list(zip(users,passwords))
print(accounts)
for account in accounts:
    telnet = telnetlib.Telnet(host)
    time.sleep(1)
    content = telnet.read_very_eager().decode('big5','ignore')
    print(content)

    if u"請輸入代號" in content:
        print("輸入帳號中...")
        telnet.write((account[0] + "\r\n").encode('ascii') )
        time.sleep(1)

        content = telnet.read_very_eager().decode('big5','ignore')

        print(content)

        if u"請輸入您的密碼" in content:
            print("輸入帳號中...")
            telnet.write((account[1] + "\r\n").encode('ascii') )
            time.sleep(1)

            content = telnet.read_very_eager().decode('big5','ignore')
    
            print(content)

            if u"請按任意鍵繼續" in content:
                print("資訊頁面，按任意鍵繼續...")
                telnet.write("\r\n".encode('ascii') )
                time.sleep(2)
                content = telnet.read_very_eager().decode('big5','ignore')
                print(content)

                telnet.write("s".encode('ascii'))
                ## telnet.write("Gossiping\r\n".encode('ascii'))
                ## time.sleep(1)
                telnet.write("\r\n".encode('ascii') )
                time.sleep(2)
                content = telnet.read_very_eager().decode('big5','ignore')
                print(content)
                telnet.close()
                time.sleep(1)
telnet.close()
```
